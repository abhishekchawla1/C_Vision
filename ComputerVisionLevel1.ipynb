```python
image=Image.open(image_bytes)
```

```python
image
```

```python
img_arr=np.array(image)
```

```python
image=cv.cvtColor(img_arr,cv.COLOR_RGB2BGR)
```

```python
plt.imshow(image)
```

```python
image_gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
```

```python
cv.imshow('grayscale',image_gray)
```

```python
plt.imshow(cv.cvtColor(image_gray,cv.COLOR_BGR2RGB))
```

```python
plt.imshow(image_gray,cmap='gray')
```

```python
image_gray.shape
```

```python
res_image=cv.resize(image_gray,(400,400))
plt.imshow(res_image,cmap='gray')
```

```python
plt.imshow(cv.flip(image_gray,1))
```

```python
plt.imshow(cv.flip(image_gray,0))
```

```python
plt.imshow(cv.flip(image_gray,-1))
```

```python
blank=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank.shape
```

```python
plt.imshow(blank)
plt.axis('off')
plt.show()
```

```python
rectangle=cv.rectangle(blank,(45,12),(100,150),thickness=3,color=(0,255,0))
```

```python
plt.imshow(rectangle)
plt.axis('off')
plt.show()
```

```python
rectangle2=cv.rectangle(rectangle,(200,200),(300,300),thickness=5,color=(255,0,0))
```

```python
plt.imshow(rectangle2)
plt.axis('off')
plt.show()
```

```python
circle=cv.circle(rectangle2,(350,400),radius=20,thickness=5,color=(0,0,255))
```

```python
plt.imshow(circle)
plt.axis('off')
plt.show()
```

```python
filled_circle=cv.circle(circle,(400,100),radius=80,thickness=-1,color=(120,42,120))
```

```python
plt.imshow(filled_circle)
plt.axis('off')
plt.show()
```

```python
line=cv.line(filled_circle,(0,500),(511,500),thickness=5,color=(102,102,21))
```

```python
plt.imshow(line)
plt.axis('off')
plt.show()
```

```python
text=cv.putText(line,org=(40,480),text='Hello To CV',fontScale=3,fontFace=cv.FONT_HERSHEY_COMPLEX_SMALL,color=(45,85,90),thickness=5)
```

```python
plt.imshow(text)
plt.axis('off')
plt.show()
```
