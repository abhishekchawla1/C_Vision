```python
result=pytesseract.image_to_data(img_rgb,lang='eng',output_type=Output.DICT)
```

```python
result
```

```python
img_copy=img_rgb.copy()
```

```python
len(result)
```

```python
len(result['text'])
```

```python
def bounding_box(result,img,i,color=(255,100,0)):
    x=int(result['left'][i])
    y=int(result['top'][i])
    width=int(result['width'][i])
    height=int(result['height'][i])
    cv.rectangle(img,(x,y),(x+width,y+height),color,2)
    text=result['text'][i]
    cv.putText(img,text=text,org=(x,y-10),fontFace=cv.FONT_HERSHEY_COMPLEX_SMALL,fontScale=5,thickness=1,color=(255,0,0))
    return x,y,img
```

```python
min_conf=50
for i in range(0,len(result['text'])):
    confidence=int(result['conf'][i])
    if confidence>min_conf:
        x,y,img_rec=bounding_box(result,img_copy,i)
plt.imshow(img_rec)
```

Searching for Specific Components

```python
import re
```

```python
img=cv.imread(r"E:\Courses\Optical Character Recognition\Images\table_test.jpg")
```

```python
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
result=pytesseract.image_to_data(img,lang='por',output_type=Output.DICT)
```

```python
result
```

```python
def bounding_box_with_conf(result,img,i,color=(255,255,0)):
    x=int(result['left'][i])
    y=int(result['top'][i])
    width=int(result['width'][i])
    height=int(result['height'][i])
    cv.rectangle(img,(x,y),(x+width,y+height),color,thickness=1)
    #text=result['text'][i]
    #conf=result['conf'][i]
    #text=text+'Confidence: '+str(conf)
    #cv.putText(img,org=(x,y-10),text=text,fontFace=cv.FONT_HERSHEY_COMPLEX_SMALL,fontScale=0.2,color=(255,25,20),thickness=1)
    return x,y,img
```

```python
len(result['height'])
```

```python
pattern=re.compile(r'^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$')
```

```python
min_conf=50
for i in range(0,len(result['height'])):
    img_copy=img.copy()
    conf=int(result['conf'][i])
    text=result['text'][i]    
    if (conf>min_conf) and (re.match(pattern,text)):
        x,y,img_box=bounding_box_with_conf(result,img,i)
        img_text=cv.putText(text=text,img=img_box,org=(x,y-5),fontFace=cv.FONT_HERSHEY_PLAIN,fontScale=0.2,thickness=1,color=(0,0,255))
plt.imshow(img_text)
```
