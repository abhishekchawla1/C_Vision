```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv.imread(r"E:\Courses\Computer-Vision-with-Python\DATA\sudoku.jpg")
```

```python
img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img,cmap='gray')
```

```python
sobel_x=cv.Sobel(img,cv.CV_64F,1,0)
sobel_y=cv.Sobel(img,cv.CV_64F,0,1)
laplacian=cv.Laplacian(img,cv.CV_64F)
```

```python
plt.imshow(sobel_x,cmap='gray')
```

```python
plt.imshow(sobel_y,cmap='gray')
```

```python
plt.imshow(laplacian,cmap='gray')
```

```python
blend=cv.addWeighted(src1=sobel_x,src2=sobel_y,alpha=0.5,beta=0.5,gamma=0)
```

```python
plt.imshow(blend,cmap='gray')
```

```python
img.shape
```

```python
blend.shape
```

```python
kernel=np.ones((4,4),np.uint8)
```

```python
grad=cv.morphologyEx(blend,cv.MORPH_GRADIENT,kernel)
```

```python
plt.imshow(grad,cmap='gray')
```

```python
kernel=np.ones((3,3),np.uint8)
```

```python
grad=cv.morphologyEx(blend,cv.MORPH_GRADIENT,kernel)
```

```python
plt.imshow(grad,cmap='gray')
```
